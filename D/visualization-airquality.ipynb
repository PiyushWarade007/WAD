```python
data=pd.read_csv(r'airquality.csv')
```

```python
#Data Cleanning
data['Ozone']=data['Ozone'].fillna(data['Ozone'].mean())
data['Solar.R']=data['Solar.R'].fillna(data['Solar.R'].mean())
data['Humidity']=data['Humidity'].fillna(data['Humidity'].mode()[0])
```
